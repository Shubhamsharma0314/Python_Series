# Dictionary -

A dictionary in Python is a built-in data structure used to store data in key–value pairs.

1. Mutable (can be changed)
2. Ordered (since Python 3.7)
3. Keys must be unique
4. Keys must be immutable (like int, string, tuple)
5. item=key + value
6. item separated by (,)
7. key- user defined index

### Creating a Dictionary -

```python
student = {
    "name": "Shubham",
    "age": 21,
    "course": "B.Tech"
}
print(student)
```

#### Access a key -

```python
d={"id":1,"name":"shubham","age":25,"salary":85000}
# d[key]
print(d["id"])
print(d["name"])
```

#### Updating the key value -

```python
d={"id":1,"name":"shubham","age":25,"salary":85000}
d["name"]="shubham sharma"
print(d)
```

#### Dynamic Dict -

```python
l=[]
l2=[]
s=int(input("Enter the size : "))
for i in range(0,s,1):
    n1=input("Enter keys : ")
    n2=input("Enter value : ")
    l.append(n1)
    l2.append(n2)
print(dict(zip(l,l2)))
```

#### Looping

### Nested dictionaries


A nested dictionary means a dictionary inside another dictionary. <br> Used when data has multiple levels (like students → subjects → marks).

```python
students = {
    "student1": {
        "name": "Shubham",
        "age": 21,
        "marks": 90
    },
    "student2": {
        "name": "Aman",
        "age": 22,
        "marks": 85
    }
}

print(students)
```

```python
print(students["student1"]["name"])   # Shubham
print(students["student2"]["marks"])  # 85
```

```python
students["student1"]["city"] = "Delhi"
print(students)
```

```python
for student_id, details in students.items():
    print("ID:", student_id)
    for key, value in details.items():
        print(key, ":", value)
    print()
```

```python
company = {
    "IT": {
        "Manager": "Raj",
        "Employees": 10
    },
    "HR": {
        "Manager": "Priya",
        "Employees": 5
    }
}

print(company["IT"]["Manager"]) 
```

### Dictionary comprehension

{key_expression: value_expression for item in iterable}

```python
squares = {x: x*x for x in range(1, 6)}
print(squares)
```

```python
marks = {"Math": 90, "Science": 85, "English": 40}

passed = {k: v for k, v in marks.items() if v >= 50}
print(passed)
```
