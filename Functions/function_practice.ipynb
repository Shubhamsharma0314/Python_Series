## Q1. Calculate Mean Function

Write a Python function called `calculate_mean` that takes a list of numbers as input and returns the mean (average). Test it with the list `[10, 20, 30, 40, 50]`.

```python
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Test
data = [10, 20, 30, 40, 50]
print("Mean:", calculate_mean(data))
```

**Output**
```
Mean: 30.0
```

## Q2. Classify Age Function

Write a function called `classify_age` that takes an age as input and returns:
- "Child" if age is below 13  
- "Teenager" if age is between 13 and 17  
- "Adult" if age is 18 and above  

```python
def classify_age(age):
    if age < 13:
        return "Child"
    elif age <= 17:
        return "Teenager"
    else:
        return "Adult"

# Test
print(classify_age(10))
print(classify_age(15))
print(classify_age(25))
```

## Q3. Default Argument

A **default argument** is a value assigned to a parameter that is used if no value is provided when calling the function.

```python
def greet_user(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# Test
print(greet_user("Shubham"))
print(greet_user("Shubham", "Welcome"))
```

**Output**
```
Hello, Shubham!
Welcome, Shubham!
```

## Q4. Using *args

Write a Python function called `describe_data` that accepts any number of numerical values using `*args` and returns the minimum, maximum, and sum.

```python
def describe_data(*args):
    return min(args), max(args), sum(args)

# Test
print(describe_data(10, 20, 5, 40, 15))
```

**Output**
```
(5, 40, 90)
```

## Q5. Filter with Lambda

Given the list of data science salaries below, use the `filter()` function with a lambda to return only salaries above 80,000.

```python
salaries = [55000, 72000, 85000, 90000, 48000, 100000, 67000]

high_salary = list(filter(lambda x: x > 80000, salaries))

print(high_salary)
```

**Output**
```
[85000, 90000, 100000]
```

## Q6. Map Function (Celsius to Fahrenheit)

Use the `map()` function to convert each value to Fahrenheit using the formula:

F = (C × 9/5) + 32

```python
Celsius = [0, 20, 37, 100]

Fahrenheit = list(map(lambda c: (c * 9/5) + 32, Celsius))

print(Fahrenheit)
```

**Output**
```
[32.0, 68.0, 98.6, 212.0]
```

## Q7. Reduce Function

Using the `reduce()` function from the `functools` module, write a program to find the product of all numbers.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

product = reduce(lambda x, y: x * y, numbers)

print(product)
```

**Output**
```
720
```

### Step-by-Step Explanation of reduce()

List: `[1, 2, 3, 4, 5, 6]`

Step 1:
```
1 * 2 = 2
```

Step 2:
```
2 * 3 = 6
```

Step 3:
```
6 * 4 = 24
```

Step 4:
```
24 * 5 = 120
```

Step 5:
```
120 * 6 = 720
```

Final Result:
```
720
```
