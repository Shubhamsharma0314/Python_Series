# Higher order functions -

- The function which takes another functions as a arguments are known as a higher order function.
1. map()
2. filter()
3. reduce()

Syntax - <br>
map(function, iteration)

```python
def demo(x):
    return x*x
```

```python
l=[1,2,3,4,5]
print(list(map(demo,l)))
```

syntax - <br>
filter(function, iteration)

```python
def even(x):
    if x%2==0:
        return x
```

```python
l=[1,2,3,4,5]
print(list(map(even,l)))
```

```python
l=[1,2,3,4,5]
print(list(filter(even,l)))
```

```python
# with lambda function - 
l=[1,2,3,4,5]
print(list(map(lambda x:x*x,l)))
```

```python
l=[1,2,3,4,5]
print(list(filter(lambda x:x%2==0,l)))
```

reduce(function, iterable)

```python
from functools import reduce
l = [1,2,3,4,5]
result = reduce(lambda a,b: a+b, l) #((1+2)+3)+4)+5
print(result)
```

```python
from functools import reduce

l = [1,2,3,4]

print(reduce(lambda a,b: a*b, l))
```

```python
from functools import reduce

l = [10,5,8,20,3]

print(reduce(lambda a,b: a if a>b else b, l))
```

map → transform each element <br>
filter → filter elements <br>
reduce → single output <br>
