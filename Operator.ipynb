# Operators -

Operators are symbols or keywords used to perform operations on variables and values or Operands. <br>
Example -> A+B <br>
Where A & B -> Operand <br>
'+' -> Operator 

1. Uninary - Operate on ONE operand
2. Binary Operators - Operate on TWO operands
3. Ternary Operator (Conditional Operator) - discuss later

+ Types of operators in Python -

```python
# Airthmetic Operator -
a = 10
b = 3
print("Addition:", a + b)          # Addition
print("Subtraction:", a - b)       # Subtraction
print("Multiplication:", a * b)    # Multiplication
print("Division:", a / b)          # Division
print("Modulus:", a % b)           # Modulus - Return the remainder
print("Exponentiation:", a ** b)   # Exponentiation - Power
print("Floor Division:", a // b)    # Floor Division -  Return the integer part of the quotient
```

```python
# assignment operator
x = 5
print("The value of X : ",x)
x += 3  # Equivalent to x = x + 3
print("Value of x after += 3:", x)
x *= 2  # Equivalent to x = x * 2
print("Value of x after *= 2:", x)
x -= 4  # Equivalent to x = x - 4
print("Value of x after -= 4:", x)
x /= 2  # Equivalent to x = x / 2
print("Value of x after /= 2:", x)
x %= 3  # Equivalent to x = x % 3
print("Value of x after %= 3:", x)
```

```python
# relational operator / comparison operator 
a = 10
b = 20
print(" a : ",a,"\n","b : ",b)
print("a == b:", a == b)  # Equal to
print("a != b:", a != b)  # Not equal to
print("a > b:", a > b)    # Greater than
print("a < b:", a < b)    # Less than
print("a >= b:", a >= b)  # Greater than or equal to
print("a <= b:", a <= b)  # Less than or equal to
```

```python
#logical operator
x = True
y = False
print("x AND y:", x and y)  # Logical AND
print("x OR y:", x or y)    # Logical OR
print("NOT x:", not x)       # Logical NOT
```

```python
# # Identity operator
a = 50
b = 50  
print(id(a))                     # Print the memory address of a
print(id(b))                     # Print the memory address of b
print("a is b:", a is b)        # Identity - True if both refer to the same object
print("a is not b:", a is not b)  # Identity - True if both refer to different objects
# for mutable objects like lists
list1 = [1, 2, 3]
list2 = list1
print(id(list1))                 # Print the memory address of list1
print(id(list2))                 # Print the memory address of list2
print("list1 is list2:", list1 is list2)  # True, both refer to the same object
print("list1 is not list2:", list1 is not list2)  # False, both refer to the same object

# for same immutable objects like strings
str1 = "hello"
str2 = "hello"
print(id(str1))                 # Print the memory address of str1
print(id(str2))                 # Print the memory address of str2
print("str1 is str2:", str1 is str2)  # True, both refer to the same object
print("str1 is not str2:", str1 is not str2)  # False, both refer to the same object

# for same lists 
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print(id(list1))                 # Print the memory address of list1
print(id(list2))                 # Print the memory address of list2
print("list1 is list2:", list1 is list2)  # False, both refer to different objects
print("list1 is not list2:", list1 is not list2)  # True, both refer to different objects
```

```python
# membership operator
my_list = [1, 2, 3, 4, 5]
print("3 in my_list:", 3 in my_list)      # Membership - True if 3 is in my_list
print("6 not in my_list:", 6 not in my_list)  # Membership - True if 6 is not in my_list
```

```python
# bitwise operator
a = 10  # In binary: 1010
b = 4   # In binary: 0100
print("a & b (AND):", a & b)   # Bitwise AND
print("a | b (OR):", a | b)    # Bitwise OR
print("a ^ b (XOR):", a ^ b)   # Bitwise XOR
print("~a (NOT):", ~a)         # Bitwise NOT
print("a << 2 (Left Shift):", a << 2)  # Left Shift
print("a >> 2 (Right Shift):", a >> 2) # Right Shift
```

```python
# walrus operator
print(x:=5)  # Assigns 5 to x and prints it
```
