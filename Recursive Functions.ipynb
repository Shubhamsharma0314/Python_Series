# Recursive Functions -

A recursive function is a function that calls itself to solve a problem.
Recursion breaks a problem into smaller subproblems until it reaches a base case.

### Factorial Using Recursion

```python
def factorial(n):
    if n == 1:     # Base case
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

### Sum of Numbers

```python
def sum_numbers(n):
    if n == 1:
        return 1
    else:
        return n + sum_numbers(n-1)
print(sum_numbers(5))
```
