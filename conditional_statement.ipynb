# Conditional Statements -

1. if-statement

```python
# Syntax -
'''
if condition1:
    # code block 1
'''
# Example -
x = 10
if x > 5:
    print("x is greater than 5")
```

2. If-else statement

```python
# syntax-
'''
if condition1:
    # code block 1
else:
    # code block 2
'''
# Example -
age=int(input("Enter your age: "))
if age>=18:
    print("your age :",age)
    print("You are eligible to vote.")
else:
    print("your age :",age)
    print("You are not eligible to vote.")
```

3. if-elif-else statement:

```python
# syntax -
'''
if condition1:
    # code block 1
elif condition2:
    # code block 2
else:
    # code block 3
'''
# Example -
num=int(input("Enter a number: "))
if num>0:
    print("number :",num)
    print("The number is positive.")
elif num==0:
    print("number :",num)
    print("The number is zero.")
else:
    print("number :",num)
    print("The number is negative.")
```

4. Nested if Statements:

```python
# syntax-
'''
if condition1:
    if condition2:
        # code block 1
    else:
        # code block 2
else:
    # code block 3
'''
# Example -
num=int(input("Enter a number: "))
if num>=0:
    if num==0:
        print("number :",num)
        print("The number is zero.")
    else:
        print("number :",num)
        print("The number is positive.")
else:
    print("number :",num)
    print("The number is negative.")
    
```

5. Match Case - 

```python
# syntax -
'''
match variable:
    case value1:
        # code block 1
    case value2:
        # code block 2
    case _:
        # default code block
'''
# Example -
day=int(input("Enter a day of the week: "))
match day:
    case 1:
        print("It's the start of the work week.")
    case 5  | 6:
        print("It's the end of the work week.")
    case _:
        print("It's just another day.")
```

6. Ternary Conditional Operator

```python
# syntax -
'''
ternary_expression = value_if_true if condition else value_if_false
''' 
# Example -
num=int(input("Enter a number: "))
result = "Even" if num % 2 == 0 else "Odd"
print("number :",num)
print("The number is", result)
```
