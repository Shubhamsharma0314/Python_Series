# Conditional Statement Practice Sheet

**Question-1.**  
Take a number from the user and check whether it is odd or even.

**Question-2.**  
Take two numbers from the user and check whether they are equal or not.

**Question-3.**  
Write a program that calculates and prints the grade based on a score.

**Question-4.**  
Write a program that classifies a person’s age into:
- Child
- Adult
- Senior Citizen

**Question-5.**  
Create a simple calculator that performs the following operations:
- Addition
- Subtraction
- Multiplication
- Division

**Question-6.**  
Write a program to check whether a given year is a leap year or not.

**Question-7.**  
Write a program for ordering food:
- Ask the user to select a type of food (Burger or Pizza)
- If Burger → ask if they want fries
- If Pizza → ask if they want extra cheese
- Print the final order with additional items (if chosen)

**Question-8.**  
Take units from the user and calculate the total electricity bill amount based on the following criteria:
- 0–200 units → Free
- 200–400 units → 5 rs per unit
- 400–800 units → 10 rs per unit
- More than 800 units → 15 rs per unit

**Question-9.**  
Write a program that asks for a user’s age and whether they have a driver’s license.  
If the user is 18 or older and has a driver’s license, print **"You are eligible to drive."**  
Otherwise, print **"You are not eligible to drive."**

**Question-10.**  
Write a Python program to allocate a stream to a student based on the following criteria:
- Marks ≥ 85 → Science
- Marks ≥ 65 and < 85 → Commerce
- Marks ≥ 45 and < 65 → Arts
