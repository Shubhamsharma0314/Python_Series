# Data Types - 

In Python, data types define the kind of value a variable can store.<br> Python is dynamically typed, meaning you donâ€™t need to declare the data type explicitly.

# There are two types of data types in Python: mutable and immutable.
- Mutable data types can be changed after they are created, while immutable data types cannot be changed.
- Examples of mutable data types: list, dict, set
- Examples of immutable data types: numbers, boolen, string, tuple,frozenset

1. Numbers - 

```python
# integer -
a=42
print(type(a))
# float -
b=3.14
print(type(b))
# complex -
c=2+3j
print(type(c))
d=None
print(type(d))
```

2. String -

```python
e="Hello, Python!"
print(type(e))
```

3. boolean -

```python
a=True
print(type(a))
```

4. list -

```python
li = [1, 2, 3, 4, 5]
print(type(li))
```

5. tuple -

```python
t=(1, 2, 3)
print(type(t))
```

# How to declare one element tuple?

```python
single_element_tuple = (5,) # Note the comma
print(type(single_element_tuple))
```

6. dictionary -

```python
d={}  # empty dictionary
print(type(d))
dict_example = {'name': 'Alice', 'age': 30}
print(type(dict_example))
```

7. set

```python
s={1,2,2,3,3}
print(type(s))
print(s)
s1={1, 2, 3}
print(type(s1))
```

8. Frozenset - 

```python
f=frozenset([1, 2, 3, 4])
print(type(f))
```
