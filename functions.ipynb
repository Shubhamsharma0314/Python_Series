# Functions in Python -

A function in Python is a reusable block of code that performs a specific task. Functions help make code modular, readable, and reusable.

### Syntax - <br>
def function_name(): <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;function_definition

```python
def greet():
    print("Namaste, Python!")
```

```python
greet() #Calling a function
```

### Types of Functions -
1. Built-in Functions-> print(), input(), etc.... <br>
2. User defined -> function definition + function calling<br>
Example - greet(),sum(),shubham()


#### Function without Argument & without return -

```python
def demo():
    print("Python Functions")
    
demo()
```

#### Function with Argument & without return -

```python
def greet(name):
    print("Namaste",name)

greet("Shubham")
```

#### Function without Argument & with return -

```python
def demo():
    return "Python Functions"
    
print(demo())
```

#### Function with Argument & with return -

```python
def greet(name):
    return "Namaste "+name

print(greet("Shubham"))
```

### Types of Arguments .....

1. Required Argument -<br>
A required argument is a parameter that must be passed when calling a function.

```python
def greet(name):
    print("Hello", name)

greet() 
```

```python
# Correct way -
greet("Shubham")
```

2. Keyword Argument -<br>
A keyword argument is when you pass arguments to a function using the parameter name = value format.

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(name="Shubham", age=21)
```

3. default Argument -<br>
A default argument is a parameter that already has a value assigned to it in the function definition.

```python
def greet(name="Unknown"):
    print("Hello", name)

greet()           
greet("Shubham")  
```

4. Arbitrary Argument -<br>
Arbitrary arguments allow a function to accept any number of arguments.

Types of Arbitrary Arguments -<br>
1. *args → for multiple positional arguments <br>
2. **kwargs → for multiple keyword arguments

```python
# *args -
def total(*numbers):
    print(numbers)
    print(sum(numbers))

total(10, 20, 30)
```

```python
def info(**data):
    print(data)

info(name="Shubham", age=21, city="Delhi")
```
