# Lambda Functions -

A lambda function is a small anonymous function (a function without a name).
It is used when you need a simple function for a short time.

Syntax : <br>
lambda arguments : expression

```python
add = lambda a, b: a + b
print(add(5, 3))
```

```python
print((lambda x, y: x + y)(10, 5)) # Lambda Without Storing
```

### Key Features -
1. No function name
2. Only one expression allowed
3. Can take multiple arguments
4. Used for short operations

### Limitations -
1. Only one expression allowed
2. Cannot contain multiple statements
3. Hard to read for complex logic
