# List
- Data Structure
- Mutable 
- Denoted by [ ]
- Store different types of data
- Indexing & slicing are possible

### Example -

```python
l=[12,"shubham",18.0,True]
print(type(l))
```

#### Indexing -

```python
students=["Ram","shyam","amit","sumit","ravi","kavi"]
print(students[0])
print(students[-1])
print(students[-3])
print(students[4])
print(students[10])
```

#### Slicing -

```python
students=["Ram","shyam","amit","sumit","ravi","kavi"]
print(students[0:4])
print(students[0:4:1])
print(students[0:4:2])
print(students[::])
print(students[0:4:-1])
print(students[::-1])
print(students[4:1:-1])
```

#### Nested List -

```python
l=[1,2,4,6,3.5,6.4,["python"]]
print(l[-1][0][0])
```

#### Iteration -

```python
students=["Ram","shyam","amit","sumit","ravi","kavi"]
for i in students:
    print(i)
```

#### append() -

```python
person_age=[12,34,56,34]
print(f'Before : {person_age}')
person_age.append(88)
print(f'After : {person_age}')
```

#### Dynamic list creation -

```python
n=int(input("Enter the size of list : "))
new_list=[]
for i in range(1,n+1):
    n1=input("Enter element : ")
    print(f'Input : {n1} Added !!')
    new_list.append(n1)
print(new_list)
```
