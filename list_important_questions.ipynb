# List Important Questions -

#### 1. Sum of all numbers in a list

```python
numbers = [10, 20, 30, 40, 50]
total = sum(numbers)  # Using built-in
print(total)

# Without built-in
total = 0
for num in numbers:
    total += num
print(total)
```

#### 2. Remove all occurrences of a value

```python
my_list = [1, 2, 3, 2, 4, 2, 5]
value_to_remove = 2

# Method 1: List comprehension
new_list = [x for x in my_list if x != value_to_remove]
print(new_list)
# Method 2: While loop
while value_to_remove in my_list:
    my_list.remove(value_to_remove)
print(my_list)
```

#### 3.  Find largest and smallest without max()/min()

```python
numbers = [45, 12, 78, 23, 90, 5]

largest = numbers[0]
smallest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print("Largest:", largest, "Smallest:", smallest)
```
