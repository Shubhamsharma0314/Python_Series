# Local & Global Variable - 

### Local Variable - <br>
A local variable is declared inside a function and can only be used within that function.
<br>
1. Created inside a function
2. Accessible only inside that function
3. Destroyed when the function finishes execution

```python
def my_function():
    x = 10   # local variable
    print("Value of x:", x)

my_function()
```

```python
def test():
    a = 5

test()
print(a)   # Error
```

### Global Variable -

A global variable is declared outside all functions and can be used anywhere in the program.

```python
x = 20   # global variable

def show():
    print("Value of x:", x)

show()
```
