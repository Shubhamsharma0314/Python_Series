# Loops -

- A loop is a programming construct that repeats a block of code multiple times until a certain condition is met.

- There are two types of loops in Python: <br>for loops and while loops.

### while loop
- A while loop repeatedly executes a block of code as long as a specified condition remains true.
<br>
- Syntax: <br>
while condition:<br>
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;code block

```python
# while loop
count = 0
while count < 5:
    print("Count is:", count)
    count += 1
```

```python
# sum of first n natural numbers-
n = int(input("Enter a positive integer: "))
sum_n = 0
i = 1
while i <= n:
    sum_n += i
    i += 1
print("The sum of the first", n, "natural numbers is:", sum_n)
```

```python
# even numbers from 1 to 20
num = 2
while num <= 20:
    print(num)
    num += 2
```

```python
# factorial of a number
number = int(input("Enter a positive integer: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print("The factorial of", number, "is:", factorial)
```

```python
# Prime number check
# method-1.
num = int(input("Enter a positive integer: "))
is_prime = True
i = 2
while i <= num // 2:
    if num % i == 0:
        is_prime = False
        break
    i += 1
if is_prime and num > 1:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
# Prime number check
# method-1.
num = int(input("Enter a positive integer: "))
is_prime = True
i = 2
while i <= num // 2:
    if num % i == 0:
        is_prime = False
        break
    i += 1
if is_prime and num > 1:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
# method-2.(faster)
num = int(input("Enter a positive integer: "))
is_prime = True
i = 2
while i <= num ** 0.5:
    if num % i == 0:
        is_prime = False
        break
    i += 1
if is_prime and num > 1:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
# method-3. (slowest)
num = int(input("Enter a positive integer: "))
is_prime = True
i = 2
while i <= num - 1:
    if num % i == 0:
        is_prime = False
        break
    i += 1
if is_prime and num > 1:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

### for loop
- Definition: A for loop iterates over a sequence (like a list, tuple, or string) and executes a block of code for each item in the sequence.
<br>
- Syntax: <br>
for variable in iterable:<br>
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;code block

```python
# for loop
for i in range(5):
    print("Iteration:", i)
    
```

```python
# sum of first n natural numbers using for loop
n = int(input("Enter a positive integer: "))
sum_n = 0
for i in range(1, n + 1):
    sum_n += i
print("The sum of the first", n, "natural numbers is:", sum_n)
```

```python
# factorial of a number using for loop
n = int(input("Enter a positive integer: "))
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print("The factorial of", n, "is:", factorial)
```

```python
# prime number check using for loop
n = int(input("Enter a positive integer: "))
is_prime = True
for i in range(2, n):
    if n % i == 0:
        is_prime = False
        break
if is_prime and n > 1:
    print(n, "is a prime number.")
else:
    print(n, "is not a prime number.")
```

### Nested Loops -
- it is a loop inside a loop.


```python
### Nested loops-
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i = {i}, j = {j}")
```

```python
# Pattern printing using nested loops-
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end="")
    print()
```

```python
# reversed pattern printing-
n = 5
for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print("*", end="")
    print()
```

```python
# abcd pattern printing-
n = 4   
for i in range(0,n+1):
    for j in range(0,i+1):
        print(chr(97 + j), end="")
    print()
```

```python
# ABCD pattern printing-
n = 4   
for i in range(0,n+1):
    for j in range(0,i+1):
        print(chr(65 + j), end="")
    print()
```
