# Loops Practice Sheet

**Question-1.**  
Write a program to find the factorial of a number using a loop.

**Question-2.**  
Reverse a string using a for loop (e.g., "hello" → "olleh").

**Question-3.**  
Count the number of vowels in a given string.

**Question-4.**  
Generate the first n numbers in the Fibonacci sequence.

**Question-5.**  
Print numbers from 10 down to 1.

**Question-6.**  
Find the sum of digits in a number (e.g., 12345 → 15).

**Question-7.**  
Find the largest number in a list without using the max() function.

**Question-8.**  
Check if a number is an Armstrong number (e.g., 153 = 1³ + 5³ + 3³).

**Question-9.**  
Check if a number is perfect (sum of divisors equals the number)

**Question-10.**  
Print all prime numbers between 1 and 100
