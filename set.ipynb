# Set

A set in Python is a collection of:<br>
1. Unique elements (no duplicates)
2. Unordered
3. Mutable (can add/remove items)
4. No indexing

```python
nums = {1, 2, 3, 4}
print(nums)
```

```python
empty = {}      # Dictionary
empty = set()   # Set
```

```python
nums = {1, 2, 3}
nums.add(4)
print(nums)
```

```python
nums.remove(2)   # Error if not present
nums.discard(5)  # No error
nums.pop()       # Removes random element
```

```python
nums.clear()
print(nums)
```

```python
s={1,2,3,4,5}
min(s)
```

```python
s={1,2,3,4,5}
max(s)
```

```python
del s
print(s)
```

# Set Operations-

### 1. Union

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)
```

### 2. Intersection 

```python
print(A & B)
```

#### 3. Difference -

```python
print(A - B)
```

### 4. Symmetric difference

```python
print(A ^ B)
```

# Set Comprehension

```python
squares = {x*x for x in range(1, 6)}
print(squares)
```

# Frozenset 

A frozenset in Python is an immutable version of a set.

```python
nums = frozenset([1, 2, 3, 4])
print(nums)
```

```python
nums = frozenset([1, 2, 3])
nums.add(4) 
```
