# String

### String features -
1. Collection of characters
2. Immutable sequence of Unicode code points
3. Enclosed in single quotes, double quotes, or triple quotes
4. Supports indexing and slicing operations

```python
s="python with shubham"
print(type(s))
```

```python
s='''python with
shubham'''
print(type(s))
print(s)
```

#### Indexing and Slicing in String -

```python
s="python with shubham"
print(s[0])          # first character
print(s[-1])         # last character
print(s[0:6])       # slicing first 6 characters
print(s[7:])        # slicing from index 7 to end
print(s[:6])        # slicing from start to index 6
print(s[-6:])       # slicing last 6 characters
print(s[:-6])       # slicing from start to -6 index
print(s[::2])       # slicing with step 2
print(s[::-1])      # reversing the string
print(s[::3])       # slicing with step 3
print(s[1:10:2])    # slicing from index 1 to 10 with step 2
print(s[-10:-1])    # slicing from -10 to -1 index
print(s[-1:-10:-1]) # slicing from -1 to -10 index in reverse order
print(s[5:15:-1])   # empty string as step is negative but start<end
print(s[15:5:-1])   # slicing from index 15 to 5 in reverse order
```

#### find(string, begin, end)- we can get the index of the any character.

```python
s="boy"
print(s.find("o"))
print(s.find("o",5))
print(s.find("z"))
```

```python
s="boy"
print(s.index("o"))
print(s.index("z"))
```

#### Data cleaning------
1. lstrip()-left strip -> remove any char from left side  
2. rstrip()- Right strip-> remove any char from left side 
3. strip()- strip from both side (right & left)

```python
d="----------------shubham************"
print(d.lstrip("-"))
print(d.rstrip("*"))
print(d.strip("-*"))
```
