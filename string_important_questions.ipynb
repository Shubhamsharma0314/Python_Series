### 1. Reverse a String

```python
s=input("Enter the string : ")
rev_string=s[::-1]
print(f'String : {s}')
print(f'Reverse string : {rev_string}')
```

### 2. Check if String is Palindrome

```python
s=input("Enter the string : ")
rev_string=s[::-1]
if s==rev_string:
    print(f'String : {s}')
    print("String is palindrome!!")
else:
    print(f'String : {s}')
    print("String is not Palindrome!!")
```

### 3. Count Vowels and Consonants

```python
s = input("Enter the string : ")

v_count = 0
c_count = 0

for i in s:
    if i.isalpha():  
        if i in "aeiouAEIOU":
            v_count += 1
        else:
            c_count += 1

print(f'Vowels : {v_count}')
print(f'Consonants : {c_count}')

```

#### 4. No. of upper case character, lower case character, no. of digit, no.  of spaces in string -   

```python
s = input("Enter the string : ")
upper_case=0
lower_case=0
special_character=0
digit=0
for i in s:
    if i.isupper():
        upper_case+=1
    elif i.islower():
        lower_case+=1
    elif i.isdigit():
        digit+=1
    else:
        special_character+=1

print(f'String : {s}')
print(f'Upper case : {upper_case}')
print(f'Lower case : {lower_case}')
print(f'Digit : {digit}')
print(f'Special Character : {special_character}')
```

### 5. Remove duplicate characters

```python
s = input("Enter the string : ")
result = ""
for ch in s:
    if ch not in result:
        result += ch
print(f'String : {s}')
print(result)
```
