# Tuple -
- A tuple is a collection data type in Python that is:
1. Ordered
2. Immutable (cannot be changed)
3. Allows duplicate values
4. denoted by ()

```python
t=(1,2,3,4,5)
t
```

```python
type(t)
```

```python
t=(1)
type(t)
```

```python
t=(1,)
type(t)
```

## Tuple Methods -

### tuple() -

```python
list=[1,2,3,4,5]
t=tuple(list)
print(t)
```

### len()

```python
len(t)
```

### sum()

```python
sum(t)
```

### min()

```python
min(t)
```

### max()

```python
max(t)
```

### index()

```python
t=("Ram","Shyam","Amit","Sumit")
t.index("Shyam")
```

### count()

```python
t=(1,1,1,2,3,4,5,5,5,3)
t.count(1)
```

### Iteration of tuple -

```python
t=(1,2,3,4,5)
for i in t:
    print(i)
```

### del 

```python
del t
print(t)
```

### Dynamic Tuple -

```python
l=[]
n=int(input("Enter no. of ele : "))
for i in range(1,n+1):
    n1=input("Element : ")
    l.append(n1)

print(tuple(l))
```
